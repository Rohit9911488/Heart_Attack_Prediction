```python
import pandas as pd
```

2.Importing the Dataset

```python
data=pd.read_csv('heart.csv')
```

3.Taking Care Of Missing Values

```python
data.isnull().sum()
```

4.Taking Care of Duplicate Values

```python
data_dup=data.duplicated().any()
```

```python
data_dup
```

5.Data Processing

```python
cat_value=[]
con_value=[]
for column in data.columns:
    if data[column].nunique()<=10:
        cat_value.append(column)
    else:
        con_value.append(column)
```

```python
cat_value
```

```python
con_value
```

6.Encoding Categorial Data

```python
data['cp'].unique()
```

```python
data.head()
```

```python
cat_value.remove('sex')
cat_value.remove('target')
data=pd.get_dummies(data,columns=cat_value,drop_first=True)
```

```python
data.head()
```

```python
data.head()
```

```python
''''
import numpy as np
data = data.replace('?', np.nan)
data = data.fillna(0 and 1)
''''
```

7.Features Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
data[con_value]=st.fit_transform(data[con_value])
```

```python
data.head()
```
